```python
import itertools
import numpy as np

features = vectorizer.get_feature_names_out()
x_instance = vectorizer.transform([docs[20]])
idx_features = x_instance.nonzero()[1]
fitur_aktif = [features[i] for i in idx_features]
print("Fitur aktif", fitur_aktif)

def predict_proba(x_vec):
    if hasattr(x_vec, "shape") and len(x_vec.shape) == 1:
        x_vec = x_vec.reshape(1, -1)
    return clf.predict_proba(x_vec)[0,1]

def pred_subset(subset_idx):
    x_masked = x_instance.toarray().flatten()
    keep = set(subset_idx)
    for j in range(len(x_masked)):
        if j not in keep:
            x_masked[j] = 0.0
    return clf.predict_proba(x_masked.reshape(1, -1))[0,1]

shap_values = {}

for target_idx in idx_features:
    kontribusi = []
    for k in range(len(idx_features)+1):
        for subset in itertools.combinations([f for f in idx_features if f!=target_idx], k):
            base_pred = pred_subset(subset)
            new_pred = pred_subset(subset + (target_idx,))
            diff = new_pred - base_pred
            kontribusi.append(diff)
    shap_values[features[target_idx]] = np.mean(kontribusi)

print("======================================================")
for f, v in shap_values.items():
    print(f"{f}: {v:.3f}")

baseline = pred_subset(())
prediksi_full = pred_subset(tuple(idx_features))

print("\nBaseline:", baseline)
print("Prediksi full:", prediksi_full)
print("Baseline + ΣSHAP:", baseline + sum(shap_values.values()))
```

```python
def predict_proba(x_vec):
    if hasattr(x_vec, "shape") and len(x_vec.shape) == 1:
        x_vec = x_vec.reshape(1, -1)
    return clf.predict_proba(x_vec)[0,1]


def predict_on_subset(subset, x_instance, clf):
    x_masked = x_instance.copy().toarray().flatten()
    keep = set(subset)
    for j in range(len(x_masked)):
        if j not in keep:
            x_masked[j] = 0.0
    return clf.predict_proba(x_masked.reshape(1, -1))[0,1]

def sample_subset(features,target,num_samples=200):
    samples = []
    for _ in range(num_samples):
        subset = [f for f in features if f!=target and random.random()<0.5]
        samples.append(subset)
    return samples

def mask_instance(subset, features):
    mask = np.zeros(len(features))
    for f in subset:
        idx = list(features).index(f)
        mask[idx] = 1
    return mask


def kernel_weight(M,k):
    if k == 0 or k == M:
        return 1e-6
    else:
        return (M-1)/scipy.special.comb(M,k)*k*(M-k)

def shap_kernel_instance(sentence, vectorizer, clf, num_samples=500):
    x_instance = vectorizer.transform([sentence])   # sparse csr
    idx_features = x_instance.nonzero()[1]
    fitur_aktif = [features[i] for i in idx_features]
    print("Fitur aktif:", fitur_aktif)

    X_mask = []
    y = []
    weights = []

    M = len(idx_features)
    total_subsets = 2**M
    print(f"Total fitur aktif = {M}, total subset = {total_subsets}")

    # ambil semua subset fitur aktif (dari 0 sampai M)
    if num_samples >= total_subsets:
        subsets = [list(s) for k in range(M+1) for s in itertools.combinations(idx_features, k)]
    else:
        subsets = []
        for _ in range(num_samples):
            k = np.random.randint(0, M+1)
            s = np.random.choice(idx_features, size=k, replace=False)
            subsets.append(list(s))

    print(f"Jumlah subset yang dipakai: {len(subsets)}\n")

    for subset in subsets:
        # hitung prediksi dengan subset ini
        pred = predict_on_subset(subset, x_instance, clf)
        y.append(pred)

        # buat mask biner
        mask = np.zeros(len(idx_features))
        for s in subset:
            idx = list(idx_features).index(s)
            mask[idx] = 1

        # kernel weight
        w = kernel_weight(M, len(subset))
        weights.append(w)

        # simpan ke X
        X_mask.append(np.concatenate(([1], mask)))

        # --- debug print ---
        subset_features = [features[i] for i in subset]
        print(f"Subset: {subset_features}")
        print(f"  Mask biner: {mask}")
        print(f"  |S|={len(subset)}, Kernel weight={w:.6f}")
        print(f"  Prediksi y={pred:.6f}\n")

    # convert ke array numpy
    X_mask = np.array(X_mask)
    y = np.array(y)
    W = np.diag(weights)

    print("=== Matriks dan Vektor ===")
    print("X_mask shape:", X_mask.shape)
    print("y shape:", y.shape)
    print("W diag shape:", W.shape, "\n")

    # Weighted Least Squares
    beta = np.linalg.pinv(X_mask.T @ W @ X_mask) @ (X_mask.T @ W @ y)
    base_value = beta[0]
    shap_values = beta[1:]

    print("=== Hasil Weighted Least Squares ===")
    print("beta (intercept + koefisien):", beta)
    print("Intercept (baseline):", base_value)
    print("Koefisien (shap values):", shap_values, "\n")

    for f, val in zip(fitur_aktif, shap_values):
        print(f"{f}: {val:.4f}")

    print("\nBaseline:", base_value)
    print("Baseline + ΣKernelSHAP:", base_value + shap_values.sum())
    print("Prediksi asli:", predict_proba(x_instance))
```

```python
import shap
import matplotlib.pyplot as plt
import numpy as np

def shap_kernel_explainer(sentence, vectorizer, clf, features, nsamples=32):
    x_instance = vectorizer.transform([sentence])
    
    def f(X):
        return clf.predict_proba(X)[:, 1]
    
    explainer = shap.KernelExplainer(f, vectorizer.transform([""]))
    
    shap_values = explainer.shap_values(x_instance, nsamples=nsamples)
    
    idx_features = x_instance.nonzero()[1]
    fitur_aktif = [features[i] for i in idx_features]
    
    print("Fitur aktif:", fitur_aktif)
    print("Hasil shap.KernelExplainer:")
    for ftr, val in zip(fitur_aktif, shap_values[0][idx_features]):
        print(f"  {ftr}: {val:.4f}")
    
    base_value = explainer.expected_value
    pred_asli = clf.predict_proba(x_instance)[0, 1]
    shap_sum = np.sum(shap_values[0])
    
    print("Base value (expected value):", base_value)
    print("Prediksi asli:", pred_asli)
    print("Baseline + ΣKernelSHAP:", base_value + shap_sum)
```
